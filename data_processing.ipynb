```python
# Separando os dados da Classe 1 e Classe 2

class1Df = df[(df['IND_BOM_1_1'] == 1) & (df['IND_BOM_1_2'] == 0)]
class2Df = df[(df['IND_BOM_1_1'] == 0) & (df['IND_BOM_1_2'] == 1)]
```

```python
# Função para separar em treinamento, validação e teste. A proporção não vai mudar então deixei constante

def splitTrainValTest(df):
    trainDf = df.sample(frac = 0.5)

    remainderDf = df.drop(trainDf.index)

    validationDf = remainderDf.sample(frac = 0.5)
    testDf = remainderDf.drop(validationDf.index)
    
    return trainDf, validationDf, testDf
```

```python
# Dividindo classe 1 em treinamento, validação e teste

class1TrainDf, class1ValidationDf, class1TestDf = splitTrainValTest(class1Df)
```

```python
# Dividindo classe 2 em treinamento, validação e teste

class2TrainDf, class2ValidationDf, class2TestDf = splitTrainValTest(class2Df)
```

```python
len(class1TrainDf) / len(class2TrainDf)
```

```python
# Função pra balancear o tamanho entre 2 dataframes de tamanhos diferentes.
# Utiliza a razão n entre tamanho dos dois df para repetir n vezes o menor df
# Ao final corta o dataframe para garantir que os dois tenham exatamente o mesmo tamanho

def balanceDf(biggerDf, smallerDf):
    difference = math.ceil(len(biggerDf) / len(smallerDf))
    for _ in range(difference):
        smallerDf = smallerDf.append(smallerDf)
        
    return smallerDf.iloc[0:len(biggerDf), :]
```

```python
# Balanceando o conjunto de treinamento

class2TrainDf = balanceDf(class1TrainDf, class2TrainDf)
```

```python
# Balanceando o conjunto de validação

class2ValidationDf = balanceDf(class1ValidationDf, class2ValidationDf)
```

```python
# Criando os conjuntos finais através da concatenação das duas classes e aleatorização do conjunto inteiro
# sample(frac=1) do pandas retorna um sample aleatorio do dataframe inteiro

trainDf = class1TrainDf.append(class2TrainDf)
trainDf = trainDf.sample(frac=1)

validationDf = class1ValidationDf.append(class2ValidationDf)
validationDf = validationDf.sample(frac=1)

testDf = class1TestDf.append(class2TestDf)
testDf = testDf.sample(frac=1)
```

```python
# Persistindo os arquivos

trainDf.to_csv('data/train.csv', sep='\t')
validationDf.to_csv('data/validation.csv', sep='\t')
testDf.to_csv('data/test.csv', sep='\t')
```
